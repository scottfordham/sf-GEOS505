I want to make a dictionary for months, since the number of days in each month is variable.
```python
alpha = 42
print(f'{alpha:01d}_{alpha:02d}_{alpha:03d}_{alpha:04d}_{alpha:05d}')
```

It looks like the 01d - 05d handles integers even when they exceed the degree specified with the 0padding syntax. 42 still prints 42 even at 01d, so this method can handle it!

Now that we have tested different list structure (especially for a single item list) and how to add zeros to our printed strings, we can loop!

```python
for year in [2025]:
 for month in range(1,13):
  for day in range(1,month_day[month] + 1): 
    for vvv in range(0,361, 3):
        if year == 2025 and month == 3 and day <= 3 and vvv in (6,120):
          print(f'{prefix}{year}_{month:02d}_{day:02d}_{vvv:03d}{suffix}')
```

I wanted to explore using the dictionary in this loop, but while I was testing, ran into an issue where the last day of each month was getting excluded. month_day[month] pulls the current iteration of month from the previous for loop and associates it with the dictionary value - I added +1 to make sure that March has 31 days instead of 30, for example.
