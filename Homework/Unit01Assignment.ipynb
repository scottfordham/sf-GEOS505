```python
def IsLeapYear(year):
    
    if (year % 4 == 0): #divisible by 4 with no remainder?
        if (year % 500 == 0): #divisible by 500 with no remainder? shouldnt this be 400? 2400 isnt divisible by 500, but is a leap year
            return True
        elif (year % 100 == 0): #leap years are absurd that they don't happen on centuries that are not divisible by 400
            return False #EX: 1900 is not a leap year because 1900/400 is 4.75
        else:
            return True
    else:
        return False # returns for ifs and elif conditionals
    
def ReturnFileList(file_base, file_ext, start_yr, end_yr):
    
    DiM = [31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31] #day in month - this is like my month_day dictionary above. I am describing it as a numbered list, this is just a vector?
    year = start_yr #initialize inputs
    files = [] #create blank vector to store output
    
    for i in range(end_yr - start_yr + 1): #calculates how many times to loop... 2000-1998 +1 = 3, this is so it is inclusive of both start and end year instead of the difference between them
        for j in range(12): #range of 12 works because DiM[0] is January
            if (j+1==2) and IsLeapYear(year): #if month is feb and leap year its 29 days, otherwise DayInMonth (DiM) is the Jth value in the vector
                days_in_month = 29
            else:
                days_in_month = DiM[j]
            
            for k in range(days_in_month): #loopable through the indexed vector that changes based on state of "J" in previous for loop
                
                file_name = f'{file_base}-{year:4d}-{(j+1):02d}-{(k+1):02d}{file_ext}' #print f-string with zero padding
                files.append(file_name) #add to the end of "files" data structure
                
        year = year + 1 #proceed to loop through next year
    
    return files #return completed list of files
```
